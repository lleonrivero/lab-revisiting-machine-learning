## Identify columns that over 85% missing values

```python
nulls_percent_df = pd.DataFrame(data.isna().sum()/len(data)).reset_index()
nulls_percent_df.columns = ['column_name', 'nulls_percentage']
#nulls_percent_df
```

```python
columns_above_threshold = nulls_percent_df[nulls_percent_df['nulls_percentage']>0.85]
columns_above_threshold['column_name']
```

```python
drop_list2 = list(columns_above_threshold['column_name'])
print(drop_list2)
```

```python
drop_list = drop_list1 + drop_list2
```

## Remove those columns from the dataframe

```python
data = data.drop(drop_list, axis=1)
```

```python
data.shape
```

## Reduce the number of categories in the column GENDER.

```python
data['GENDER'].value_counts()
```

```python
data['GENDER'] = data['GENDER'].fillna('F')
```

```python
def gender(x):
    if 'F' in x:
        return 'F'
    elif 'M' in x:
        return 'M'
    else:
        return 'other'

data['GENDER'] = list(map(gender, data['GENDER']))
```
